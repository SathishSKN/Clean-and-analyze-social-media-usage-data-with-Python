<h1><center>Clean and analyze social media usage data with Python</center></h1>

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

### Creating a List

```python
categories = ["Food", "Travel", "Fashion", "Fitness", "Music", "Culture", "Family","Health"]
categories
```

### Creating a dictionary

```python
data = {'Date': pd.date_range('2023-01-01', periods=500), 'Category': [random.choice(categories) for _ in range(500)], 'Likes': np.random.randint(0, 10000, size=500)}
data
```

### Converting into Pandas DataFrame

```python
df = pd.DataFrame(data)
df.head()
```

### DataFrame Information

```python
df.info()
```

### DataFrame Description

```python
df.describe()
```

### Count of Category Variable

```python
df.value_counts('Category')
```

### Dropping Null values in DataFrame

```python
df.dropna(inplace = True)
df
```

### Dropping Duplicated values in DataFrame

```python
df.drop_duplicates(inplace = True)
df
```

### Conversion of DataFrame's Date 

```python
pd.to_datetime(df['Date'])
df
```

### Conversion of Datatype 

```python
df = df.astype({'Likes':'int'})
df.dtypes
```

### Visualizing the DataFrame

```python
sns.histplot(df['Likes'], color = 'g')
plt.title('Likes Histogram')
plt.show()
```

```python
sns.boxplot(x = 'Category', y = 'Likes', data = df)
plt.title('Boxplot of Category & Likes')
plt.show()
```

### Statistical measure of DataFrame

```python
df['Likes'].mean()
```

```python
category_likes = df.groupby('Category')['Likes'].mean()
category_likes
```
